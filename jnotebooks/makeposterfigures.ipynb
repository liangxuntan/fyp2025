```python
import matplotlib.pyplot as plt
import numpy as np

# Set random seed for reproducibility
np.random.seed(0)

# Generate random points for each group
group1 = np.random.randn(30, 2) + np.array([2, 2])   # Red group centered at (2,2)
group2 = np.random.randn(30, 2) + np.array([-2, -2]) # Blue group centered at (-2,-2)
group3 = np.random.randn(30, 2) + np.array([2, -2])  # Green group centered at (2,-2)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(group1[:, 0], group1[:, 1], color='red', label='Group 1 (Red)')
plt.scatter(group2[:, 0], group2[:, 1], color='blue', label='Group 2 (Blue)')
plt.scatter(group3[:, 0], group3[:, 1], color='green', label='Group 3 (Green)')

plt.title("")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


plt.axis('equal')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate some synthetic data
np.random.seed(42)
genome_size = np.random.normal(loc=4, scale=0.5, size=100)  # Genome size in Mbp
metabolic_preference = 0.5 * genome_size + np.random.normal(scale=0.3, size=100)  # Fake linear-ish relation

# Create a DataFrame
data = pd.DataFrame({
    'Genome Size (Mbp)': genome_size,
    'Predicted Metabolic Preference': metabolic_preference
})

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Genome Size (Mbp)', y='Predicted Metabolic Preference', s=60)
sns.regplot(data=data, x='Genome Size (Mbp)', y='Predicted Metabolic Preference', scatter=False, color='red', label='Trend')
plt.title('')
plt.xlabel('Genome Size (Mb) ')
plt.ylabel('Glucose use ')
#plt.legend()
#plt.grid(True)
plt.tight_layout()
plt.show()
```
